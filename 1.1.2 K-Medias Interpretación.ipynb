```python
# Cargar datos originales
originales = extrae_np(archivos_fuente["sponge1"], -1)
cero = originales[0]
rens = cero.shape[0]
cols = cero.shape[1]
```

```python
import matplotlib.pyplot as plt
```

```python
def np_a_img(datos3D):
    """
    Recibe la imagen del cuadro con 6 canales y regresa
    una vista RGB
    """
    return datos3D[:,:,[2,1,0]].astype(int)
```

```python
def muestra(datos3D, estimador_k_medias):
    """
    Recibe la imagen aún con componentes IXY RGB y HSV
    y el estimador tipo k-medias ya entrenado.
    """
    img = np_a_img(datos3D)
    centros = estimador.cluster_centers_[:,[1,2,5,4,3]] # i,j,R,G,B
    print(centros)
    plt.scatter(centros[1], centros[0], c=centros[3])
    plt.imshow(img)
```

```python
muestra(originales[0], estimador)
```

```python
def muestra_segmentación(ind):
    datos3D = originales[ind]
    img = np_a_img(datos3D)
    tabla = convierte_a_tabla(datos3D)
    agrupación = estimador.predict(tabla)
    etqs = agrupación.reshape((rens, cols))
    fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2)
    ax0.imshow(img)
    ax1.imshow(etqs)
```

```python
muestra_segmentación(1)
```

```python
muestra_segmentación(35)
```
