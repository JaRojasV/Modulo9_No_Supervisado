## Agrupamiento

[sklearn.cluster.KMeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```python
from sklearn.cluster import KMeans
```

```python
# Preparación del algoritmo
k = 4    # número de grupos (clústeres)
kmeans = KMeans(init="k-means++", n_clusters=k, random_state=0)
kmeans
```

```python
# Ejecución sobre el conjunto de datos
estimador = kmeans.fit(datos_1)
```

```python
# Resultados
estimador.cluster_centers_
```
